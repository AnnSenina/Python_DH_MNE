# Новый и прекрасный цикл for

```python
c = [1, 16541, 35164, 66.44538] # давайте напечатаем по 1 элементу списка в каждый строке...
l_0 = 0
l = len(c)
while l > l_0:
  print(c[l_0])
  l_0 = l_0 + 1 # чтобы цикл остановился

# while - долго и много лишних переменных
```

```python
c = [1, 16541, 35164, 66.44538]
for i in c: # i - каждый элемент в списке с по порядку
  print(i)

# быстро и красиво: цикл остановится сам, как только кончится список (или др. контейнер)
```

```python
print('a' in 'abc') # проверка, есть ли элемент в строке
print(5000 in c) # проверка, есть ли элемент в списке
```

```python
for i in c:
  print(100000000 + i)
  # инструкции менять список ПОКА не было!
  # возьми число из списка и напечатай, что с ним будет, если прибавить + 100000000
```

```python
c
# исходный список не изменился, давайте учиться делать замену в списке циклом for
```

У каждого списка есть индексы - можно заставить пайтон напечатать все индексы списка:

```python
list(range(len(c))) # генерирует индексы для всех элементов списка
```

```python
for i in range(len(c)):
  # range - генерирует последовательность чисел
  # len - считатет длину списка с
  print(i)

# нумерация - снова питонская! От нуля и без верхней границы, как в строках!
```

```python
# объединим
for i in range(len(c)): # i = 0, 1, 2, 3
  print(c[i], 'печатаем элемент')
  print(i, 'печатаем его индекс')
```

```python
# а зачем перебирать поэлементно?
# несколько типичных ситуаций:

# 1. давайте заменим все вещественные числа - целыми

# пытаетмся:
for i in c:
  if type(i) == float:
    i = round(i) # round - коварное округление, но поговорим о нем в другой раз
    print(i)
```

```python
print(c)
# замены не произошло, почему?
```

```python
# сравните:
for i in range(len(c)):
  if type(c[i]) == float:
    c[i] = round(c[i])
print(c)

# почему range разрешает менять исходный список:
# в отличие от ситуации выше, мы не просто "достаем" элементы из списка (чемодана, полки...) и как-то преобразуем
# мы обращаемся к каждому элементу по индексу и прямо делаем замену в самом списке
```

```python
# 2 типичная ситуация: несколько списков одинаковой длины

names = ['ann', 'iliana']
marks = [7, 8]
for i in range(len(marks)):
  print('Студент', names[i], 'получил оценку:', marks[i])

# а зачем здесь range?
# без range мы один и тем же циклом for могли бы перебирать ТОЛЬКО ОДИН список за раз
# с помощью range (раз длина одинакова по условию) можно обращаться к обоим спискам разом, поиндексно (по порядку!)
```

### range(от, до, шаг)

```python
# а range откуда пришел? он не страшный :)
print(*range(10)) # от 0 по умолчанию до 9

print(*range(1, 10)) # указать: от 1 до 9

d = list(range(10, 50)) # от 10 до 49 и преврати в список
print(d)
# генерирует числовые последовательности

print(list(range(0, 100, 5)), sep=', ') # от 0 до 99 с шагом в 5
# даже умеет делать это с шагом, как в индексации строк

# все генерирутеся по-питонски: от 0 по умолчанию и выбрасывая верхнуюю границу
# позволяет с помощью range(len(список)) сгенерировать все ИНДЕКСЫ элементов списка
```

# Методы

Это как бы следующая тема, но в связи со списками, нам нужны split и append

### Систаксис методов:

переменная.метод(аргументы)

переменная.сделайчто-то(как)

```python
# из строки можно получить список
text = 'hello, how are you, nice to meet you'
print(text.split())
print(text.split(', '))
# в результате деления получается список!

# строка.раздели(чем_делить)
```

```python
new_list = [] # пустой список
new_list.append(200) # добавь в конец списка число 200
print(new_list)

for i in range(10):
  print(i) # печатай
  new_list.append(i) # добавляй в конец списка очередной элемент...

print(new_list)

# список.добавь_в_конец(что_добавить)
```
